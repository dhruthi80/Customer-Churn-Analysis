```python
# in Senior citizen column there is 0 or 1 in order for bttr understanding where there is 1 then yes or else 0
df["SeniorCitizen"] = np.where(df["SeniorCitizen"] == 1,"Yes","No")
df.head(30)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (5,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels = gb.index,autopct = "%1.2f%%",colors = ("ligtgreen","red"))
plt.title("Percentage of Churned Customers",fontsize = 12)

plt.show()
```

From the given piechart we can conclude that26.54% of our customers are churned out

```python
plt.figure(figsize = (3,3))
sns.countplot(data = df, x = "gender",hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Basic countplot — only x, not hue
ax = sns.countplot(data=df, x="SeniorCitizen", palette={"No": "lightgreen", "Yes": "red"})

# Add labels to both bars
for container in ax.containers:
    ax.bar_label(container)

plt.title("Count of Customers by Churn")
plt.show()
```

```python
# Create % table
data = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot
ax = data.plot(kind='bar', stacked=True, color=['lightgreen', 'red'])

# Show % labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center', fontsize=9)

plt.title("Churn by Senior Citizen")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.tight_layout()
plt.show()

```

  Conclusion : A higher percentage of customers in the senior citizen category have churned compared to non-senior customers.

```python
plt.figure(figsize = (9,4))
sns.histplot(data = df, x = "tenure",bins = 72,hue = "Churn")
plt.show()
```

Conclusion : Customers who have used our services for a longer duration have stayed, whereas those with a tenure of just 1 or 2 months have churned

```python
ax = sns.countplot(x='Contract', data=df,hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

Conclusion :Customers with month-to-month contracts are more likely to churn compared to those with one- or two-year contracts

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

churn_colors = {"No": "lightgreen", "Yes": "red"}

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(cols):
    ax = axes[i]
    sns.countplot(x=col, data=df, hue='Churn', palette=churn_colors, ax=ax)
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=0)  # horizontal labels
    
    # Add data labels for all bar containers in this subplot
    for container in ax.containers:
        ax.bar_label(container)

plt.tight_layout()
plt.show()
```

Conclusion : The majority of customers who do not churn tend to have services like PhoneService, InternetService (particularly DSL), and OnlineSecurity enabled. For services like OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.

```python
churn_colors = {"No": "lightblue", "Yes": "yellow"}

ax = sns.countplot(x='PaymentMethod', data=df, hue='Churn', palette=churn_colors)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xticks(rotation=45) 
plt.title("Churn of Customers by Payment Method")
plt.tight_layout()
plt.show()
```
